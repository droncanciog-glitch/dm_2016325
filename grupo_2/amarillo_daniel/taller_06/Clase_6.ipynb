* Grabaciones de audio de audiencias judiciales de la Rama Judicial

    a) No-Estructurada
    b) Requiere preprocesamiento
    c) Con las librerias tuneR y Tibble ejecutar

```
library(tuneR)
library(tibble)

# 1. Read the wave file
wav <- readWave("path/to/your/audio.wav")

# 2. Convert to a tibble (left and right channels)
audio_tibble <- tibble::as_tibble(wav@left)

# Optionally, add a time column based on sample rate
audio_tibble$time <- seq(1, length(wav@left)) / wav@samp.rate

print(audio_tibble)
```  
* Factura electrónica emitida por la DIAN (formato XML)
  
    a) Semi-Estructurada

    b) Al estar en formato XML no tiene una estructura definida

    c) Con las librerias tidyr y xml2 ejecutar

```
library(xml2)
library(tidyr)
library(tibble)

# 1. Leer el archivo XML
xml_data <- read_xml("your_file.xml")

# 2. Convertir a una lista y luego a un tibble
xml_list <- as_list(xml_data)
xml_tibble <- as_tibble(xml_list) %>%
  unnest_longer(root_node_name)
```
