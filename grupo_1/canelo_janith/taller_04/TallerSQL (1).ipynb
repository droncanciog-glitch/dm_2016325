```python
import urllib.request
import os
import sqlite3
import pandas as pd
```

```python
url_chinook = "https://raw.githubusercontent.com/lerocha/chinook-database/master/ChinookDatabase/DataSources/Chinook_Sqlite.sqlite"
ruta_db     = "chinook.db"

if not os.path.exists(ruta_db):
    urllib.request.urlretrieve(url_chinook, ruta_db)
    print(f"Base de datos descargada en: {ruta_db}")
else:
    print(f"La base de datos ya existe: {ruta_db}")


```

```python
con = sqlite3.connect("chinook.db")
cur = con.cursor()
```

1. ¿Cuántas pistas hay en total en la base de datos?

```python

resultado = pd.read_sql_query("SELECT COUNT(TrackId) from Track", con)

print(f"Pistas totales: {resultado}")
```

2. ¿Cuántos géneros distintos existen?

```python
resultado = pd.read_sql_query("SELECT DISTINCT COUNT(GenreID) from Genre", con)
print(f"Generos distantos: {resultado}")
```

3. Obtén el nombre del género, el número de pistas y la duración promedio (en minutos) para cada género. Ordena de mayor a menor número de pistas.

```python
resultado = pd.read_sql_query("""
  SELECT
   DISTINCT g.name AS "genre",
   COUNT(t.TrackID) AS "total_tracks",
   AVG(t.Milliseconds / 60000) AS "duration"
  FROM Track t
  JOIN Genre g ON g.GenreID=t.GenreID
  GROUP BY g.name
  ORDER BY total_tracks DESC""", con)

resultado.head(5)
```

4. ¿Cuáles son las 5 pistas más largas? Muestra el nombre, el álbum y la duración en minutos.

```python
resultado = pd.read_sql_query("""
SELECT t.name, t.Milliseconds, a.Title
FROM Track t
JOIN Album a ON a.AlbumID=t.AlbumID
ORDER BY
Milliseconds DESC
LIMIT 5""", con)

resultado.head(10)
```

¿Cuántas pistas tienen un UnitPrice superior a $0.99? ¿Qué porcentaje representan del total?

```python
resultado = pd.read_sql_query("""
SELECT
COUNT(TrackID) AS totales,
COUNT(TrackID) * 100 / (SELECT COUNT(TrackID) FROM Track) percentaje
FROM Track
WHERE UnitPrice > 0.99
""", con)

resultado.head(10)
```

Calcula la media, mínimo, máximo y varianza (en SQL) de la duración en milisegundos de todas las pistas.

```python
resultado = pd.read_sql_query("""
SELECT
ROUND(AVG(Milliseconds) AS median,
MIN(Milliseconds) AS min,
MAX(Milliseconds) AS max,
ROUND(AVG(Milliseconds * Milliseconds) - AVG(Milliseconds)*AVG(Milliseconds), 2) AS varianza
FROM Track
""", con)


resultado.head(5)
```

Ejercicio 2: Análisis descriptivo de ventas
Usando las tablas Invoice e InvoiceLine, responde:

¿Cuál es el total de ingresos generados por la tienda? ¿Y el promedio por factura?

```python
resultado = pd.read_sql_query("""
SELECT SUM(Total),
AVG(Total) AS Mean
FROM Invoice
""", con)


resultado.head(5)
```

¿Cuántas facturas se emitieron por año? Ordena cronológicamente.

```python
resultado = pd.read_sql_query("""
SELECT strftime('%Y', InvoiceDate) AS Year,
COUNT(InvoiceID) AS Total
FROM Invoice
GROUP BY Year
ORDER BY Year
""", con)


resultado.head(5)
```

Identifica los 5 países con más ingresos. Muestra: país, número de facturas, ingreso total y promedio por factura.
